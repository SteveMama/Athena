```python
import pandas as pd

def analyze_large_file(file_path, nrows=5):
    try:
        sample_df = pd.read_csv(file_path, sep='\t', nrows=nrows)
        return sample_df.columns.tolist()
    except Exception as e:
        return f"Error: {str(e)}"

file_path = '100000-Patients/LabsCorePopulatedTable.txt'
columns = analyze_large_file(file_path)

print("Columns in the file:")
for i, col in enumerate(columns, 1):
    print(f"{i}. {col}")
```

```python
def get_dataframe(file_path):
    get_df = pd.read_csv(file_path, sep='\t',nrows=500 )
    return get_df
```
